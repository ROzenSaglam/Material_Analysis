RHEOLOGY PLOTTER CLASS

```python
import pandas as pd
import matplotlib.pyplot as plt


class RheologyPlotter:
    def __init__(self, file_path):
        self.file_path = file_path
        self.df = None
    
    def load_data(self, sheet_name):
        excel_file = pd.ExcelFile(self.file_path)
        self.df = pd.read_excel(excel_file, sheet_name)
    
    def plot_amplitude_sweep(self, save_file_name, x_col, y1_col, y2_col, x_label, y_label):
        x = self.df[x_col]
        y1 = self.df[y1_col]
        y2 = self.df[y2_col]
        
        plt.plot(x, y1, label="G' (Storage Modulus)", color="r", marker="v")
        plt.plot(x, y2, label="G'' (Loss Modulus)", color="b", marker="v")
        plt.yscale("log")
        plt.rcParams["font.family"] = "Times New Roman"
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.legend()
        plt.savefig(save_file_name, dpi=300, bbox_inches="tight")
        plt.show()
    
    def plot_frequency_sweep(self, save_file_name, x_col, y1_col, y2_col, x_label, y_label):
        x = self.df[x_col]
        y1 = self.df[y1_col]
        y2 = self.df[y2_col]
        
        plt.plot(x, y1, label="G' (Storage Modulus)", color="r", marker="v")
        plt.plot(x, y2, label="G'' (Loss Modulus)", color="b", marker="v")
        plt.yscale("log")
        plt.scale("log")
        plt.rcParams["font.family"] = "Times New Roman"
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.legend()
        plt.savefig(save_file_name, dpi=300, bbox_inches="tight")
        plt.show()

    def plot_time_sweep(self, save_file_name, x_col, y1_col, y2_col, x_label, y_label):
        x = self.df[x_col]
        y1 = self.df[y1_col]
        y2 = self.df[y2_col]
        
        plt.plot(x, y1, label="G' (Storage Modulus)", color="r", marker="v")
        plt.plot(x, y2, label="G'' (Loss Modulus)", color="b", marker="v")
        plt.yscale("log")
        plt.xscale("log")
        plt.rcParams["font.family"] = "Times New Roman"
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.legend()
        plt.savefig(save_file_name, dpi=300, bbox_inches="tight")
        plt.show()

```

```python
# Usage example
rheology_plotter = RheologyPlotter("Kitap3.xlsx")

# Amplitude sweep test for Enzymatic Levan hydrogel (EStrain)
rheology_plotter.load_data("EStrain")
rheology_plotter.plot_amplitude_sweep(
    "AmplitudeSweep20.png",
    "Shear Strain",
    "Storage Modulus",
    "Loss Modulus",
    "Strain(%)",
    "G' & G''(Pa)"
)
```

```python
# Frequency sweep test for PAFreq
rheology_plotter.load_data("PAFreq")
rheology_plotter.plot_frequency_sweep(
    "FreqSweepPA.png",
    "Frequency",
    "Storage Modulus",
    "Loss Modulus",
    "Frequency (Hz)",
    "G' & G''(Pa)"
)
```
